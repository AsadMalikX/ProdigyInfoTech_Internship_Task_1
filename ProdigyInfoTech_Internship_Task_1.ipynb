###Task-01: Implement a linear regression model to predict the prices of houses based on their square footage and the number of bedrooms and bathrooms.

### Import Libraries:

```python
# Importing important libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Load the training dataset
train_data = pd.read_csv('/content/train.csv')
```

```python
# Select relevant columns for training data
features = ['LotArea', 'BedroomAbvGr', 'FullBath', 'OverallQual', 'GrLivArea', 'YearBuilt']  # Add more relevant features
X_train = train_data[features]
y_train = train_data['SalePrice']
```

```python
# Build and train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions on the training data
y_pred_train = model.predict(X_train)
```

```python
# Evaluate the model using training data
mse_train = mean_squared_error(y_train, y_pred_train)
rmse_train = np.sqrt(mse_train)
r_squared_train = model.score(X_train, y_train)

print(f'Mean Squared Error on Training Data: {mse_train}')
print(f'Root Mean Squared Error on Training Data: {rmse_train}')
print(f'R-squared on Training Data: {r_squared_train}')
```

```python
# Plot actual vs. predicted prices for training data
plt.figure(figsize=(10, 6))
plt.scatter(y_train, y_pred_train, alpha=0.5)
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=2)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices on Training Data')
plt.show()
```
